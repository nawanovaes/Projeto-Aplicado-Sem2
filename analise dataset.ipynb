# Criando df por fonte

```python
fonte = pd.DataFrame(data.fonte.unique(), columns = ['fonte'])
fonte = fonte.fillna('-')
data['fonte'] = data['fonte'].fillna('-')
```

```python
fonte = list(fonte['fonte'])
fonte
```

```python
dfs = list(range(7))
```

```python
for i in range(len(fonte)):
    dfs[i] = pd.DataFrame(data = data[data['fonte'] == fonte[i]].replace(0, np.nan).dropna(axis = 1, how = 'all'))
```
