# Criando coluna fonte

```python
data[data['country'].str.contains('\(')].country.unique()
```

```python
data['fonte'] = data['country'].str.extract('.*\((.*)\).*')
```

# Criando coluna sem fonte

```python
data.insert(1, 'country2', data['country'])
data
```

```python
data['country2'] = data['country'].str.replace(r"\(.*\)", '').str.strip()
```

```python
data[data['country2'].str.contains('Europe')].country2.unique()
```
